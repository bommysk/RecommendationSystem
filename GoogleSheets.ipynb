```python
import pandas as pd
import numpy as np
import time
import datetime

data = pd.read_csv("https://docs.google.com/spreadsheet/ccc?key=1oWAhVLyneVU9Jt2vHHpaYTv1Clht1CpIJOmkD-tlDmk&output=csv")
movies = pd.read_csv("ml-100k/u.item", sep='|', header=None, encoding = "ISO-8859-1")

# min id value to assign google form users
beg_user_id = 944
```

```python
def timestampToUnix(s):
    return time.mktime(datetime.datetime.strptime(s, "%m/%d/%Y %H:%M:%S").timetuple())
```

```python
movies = movies.iloc[:,0:2]
movies.columns = ["item_id", "movie"]

data['user_id'] = data.index
data['user_id'] = data['user_id'].apply(lambda x: x + beg_user_id)
data.rename(columns={'Timestamp': 'timestamp'}, inplace=True)
data['timestamp'] = data['timestamp'].map(timestampToUnix)
users = data[['Please provide your name so we can evaluate the quality of your recommendations', 'user_id']].copy(deep=True)
users.rename(columns={'Please provide your name so we can evaluate the quality of your recommendations': 'name'}, inplace=True)

del data['Please provide your name so we can evaluate the quality of your recommendations']
```

```python
ratings = pd.DataFrame(columns=['movie', 'user_id', 'rating', 'timestamp'])

# iterate through each users ratings
for index, row in data.iterrows():
    
    user = row['user_id']
    
    # iterate over all movie ratings from a single user
    for x in range(0, len(row)):
        # ignore the user_id column, don't use it as a rating
        if(data.columns[x] not in ["user_id", "timestamp"] and pd.notnull(row[x])):
            # add information to the ratings table
            rating = pd.DataFrame({"movie": [data.columns[x]], "user_id": [row['user_id']], "rating": [row[x]], 'timestamp': [row['timestamp']]})
            ratings = ratings.append(rating)
ratings['user_id'] = ratings['user_id'].astype(int) 

# join item_id on ratings
ratings = ratings.merge(movies, how='inner', on='movie')
del ratings['movie']
```
